```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#function to draw deletion size distribution graph
def draw_deletion_size_distribution(sample, df):
    df.sort_values(by='Deletion Size', ascending=False, inplace=True)
    y_values = range(len(df))
    conditions = [
        (df["Deletion Size"] > 0) & (df["Deletion Size"] <= 10),
        (df["Deletion Size"] > 10) & (df["Deletion Size"] <= 50),
        (df["Deletion Size"] > 50) & (df["Deletion Size"] <= 100),
        (df["Deletion Size"] > 100)
    ]
    # Define the color for each condition
    colors = ['#53D2DC', '#3196E2', '#FF826C', '#FFC05F']
    # Apply the conditions to the 'color' column
    df['color'] = np.select(conditions, colors, default='black')  # 'black' is a default color if none of the conditions are met
    sns.set(style="whitegrid")
    title_fontsize = 14
    axis_label_fontsize = 12
    tick_label_fontsize = 10
    plt.figure(figsize=(10, 10))
    for y, (index, row) in zip(y_values, df.iterrows()):
        plt.hlines(y=y, xmin=-row['Left Deletions'], xmax=row['Right Deletions'], color=row['color'], linewidth=0.5)
    plt.axvline(x=0, color='red', linestyle='--', linewidth=1)
    plt.xlabel('Reference Position', fontsize = axis_label_fontsize)
    plt.ylabel('Number of Reads', fontsize = axis_label_fontsize)
    plt.xlim(-150, 150)
    plt.xticks(np.arange(-150, 150, 20), fontsize = tick_label_fontsize)
    plt.yticks(fontsize = tick_label_fontsize)
    plt.title(sample)
    plt.savefig(f"{sample}.jpg", dpi=300)
    
def get_files_with_pattern(directory, pattern):
    matching_files = []
    for file in os.listdir(directory):
        if file.endswith(pattern):
            matching_files.append(os.path.join(directory, file))
    return matching_files

def draw_mh_size_distribution(sample, df):

    counts = df['Microhomology'].str.len().value_counts()
    total_occurrences = counts.sum()
    length_percentages = (counts / total_occurrences) * 100
    # Create the histogram using Matplotlib
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

    axes[0].bar(counts.index, counts.values, color='skyblue', edgecolor='black')
    axes[0].set_xticks(range(1, 8))
    axes[0].set_xlabel("Microhomology Size (bp)")
    axes[0].set_ylabel('Frequency')
    axes[0].set_xlim(0,8)
    
    axes[1].bar(length_percentages.index, length_percentages.values, color='skyblue', edgecolor='black')
    axes[1].set_xticks(range(1, 8))
    axes[1].set_yticks(range(0, 101, 20))
    axes[1].set_xlabel("Microhomology Size (bp)")
    axes[1].set_ylabel('Percentage')
    axes[1].set_xlim(0,8)

    plt.suptitle(sample + " Microhomology")
    plt.grid(True)
    plt.savefig(f"{sample}_MH.jpg", dpi=300)
    
# Example usage:
directory_path = "/gpfs/home/asun/scarmapper_in_202404/igh_8_8"
file_pattern = "_Raw_Data.txt"  # Change this to your desired file pattern
matching_files = get_files_with_pattern(directory_path, file_pattern)
indexDic = pd.read_csv("/gpfs/home/asun/scarmapper_in_202404/sampleManifest.txt", header=None, sep="\t")

# Iterate over file list
for filename in matching_files:
    index = filename.split("_")[5] + "_" + filename.split("_")[6]
    rowDic = indexDic[indexDic[0] == index].index
    sample = indexDic.loc[rowDic, 1].values[0]
    print(index)
    print(sample)
    print(filename)
    
    # Read data
    df = pd.read_csv(filename, skiprows=6, sep="\t")
    print(sample, "#row(aka total#read):", len(df))
    #draw_deletion_size_distribution(sample, df)
    draw_mh_size_distribution(sample, df)
```
