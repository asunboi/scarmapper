```python
##python 20221003igh2.py input.{xx:xx}.sam groupNumber cutSite ref
##this program take the input sam files, plot deletion size for all events, and generate rawData excel files
## rawData excel include indel start position, indel size, left and right deletion size, but no MH information


import sys
import os
import re
import pandas as pd
import pysam
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from Bio import SeqIO

#read sam files, store readID, reference name, start position, CIGAR and sequence into new df
#input file path, output df
def read_sam_file(file_path):
    samfile = pysam.AlignmentFile(file_path, "rb")
    data = []
    for read in samfile:
        data.append([read.query_name, read.reference_name, read.reference_start, read.cigarstring, read.query_sequence])
    df = pd.DataFrame(data, columns=['ReadID', 'ReferenceName', 'StartPosition', 'CIGAR', 'Sequence'])
    samfile.close()
    return df


#read reference fasta file store as string
def getRef_fasta(filepath):
    with open(filepath, "r") as file:
        record = next(SeqIO.parse(file, "fasta"))
        return str(record.seq)


#function to add indel position and indel size to df
def find_indel_positions_and_sizes(cigar, start_position, cut_site, sequence, reference):    
    # parse CIGAR string into number+letter group
    elements = re.findall(r'(\d+)([MID])', cigar)
    elements = [(int(num), op) for num, op in elements]
    # CIGAR as "xM", no indel
    if len(elements) == 1 and elements[0][1] == 'M':
        return pd.Series([0, 0], index=['indel_position', 'indel_size'])
    current_position = start_position + 1
    #indel_lower_bound = cut_site - 120
    #indel_upper_bound = cut_site + 120
    indels = []
    for num, op in elements:
        if op == 'M' :
            current_position += num
        elif op == 'D':
            if current_position + num >= cut_site and current_position <= cut_site:
                indels.append((current_position, -num))  # Negative size for deletion
            current_position += num
        elif op == 'I':
            if current_position + num >= cut_site and current_position <= cut_site:
                indels.append((current_position, num))  # Positive size for insertion
        #handle cases when cigar contain soft clipping, apply for bwa alignment, ignore for bbmap alignment
        '''
        elif op == 'S':
            if num < 10:
                return pd.Series([0, 0], index=['indel_position', 'indel_size'])
            else:
                match = re.match(r'(\d+)M(\d+)S$', cigar) or re.match(r'(\d+)S(\d+)M$', cigar)
                if match:
                    seq1_len, seq2_len = map(int, match.groups())
                    seq1 = sequence[:seq1_len]
                    seq2 = sequence[-seq2_len:]
                    seq1_position = reference.find(seq1)
                    seq2_position = reference.find(seq2)
                    if seq1_position != -1 and seq2_position != -1:
                        del_start = seq1_position + seq1_len
                        del_size = seq2_position - del_start
                        return pd.Series([del_start, -del_size], index=['indel_position', 'indel_size'])
                    else:
                        return pd.Series([np.nan, np.nan], index=['indel_position', 'indel_size'])
        '''
    #indels = [(pos, size) for pos, size in indels if indel_lower_bound <= pos <= indel_upper_bound]
    if len(indels) == 0:
        return pd.Series([0, 0], index=['indel_position', 'indel_size'])
    elif len(indels) != 1:
        return pd.Series([np.nan, np.nan], index=['indel_position', 'indel_size'])
    return pd.Series(indels[0], index=['indel_position', 'indel_size'])


#function to calculate indel size percentage and store percentage in a df
def calculate_indel_size_percentage(df):
    indel_sizes = df['IndelSize']
    indel_size_percentages = indel_sizes.value_counts(normalize=True, sort=False, dropna=False) * 100
    indel_size_counts = indel_sizes.value_counts(sort=False, dropna=False)
    indel_size_df = pd.DataFrame({
        'IndelSize': indel_size_percentages.index,
        'Percentage': indel_size_percentages.values,
        'ReadCount': indel_size_counts[indel_size_percentages.index].values
    })
    indel_size_df = indel_size_df.sort_values(by='Percentage', ascending=False).reset_index(drop=True)
    return indel_size_df

def adjust_MHposition(indelSize, MHsize, LDel, RDel):
    if pd.isna(indelSize) or pd.isna(MHsize):
        return pd.Series([pd.NA, pd.NA, pd.NA],index=['adjust', 'leftDelAdjust', 'rightDelAdjust'])
    if (indelSize < 0) & (MHsize > 0):
        adjust = abs(MHsize) / 2  # Corrected: Use abs() function
        ldeladjust = LDel - adjust
        rdeladjust = RDel + adjust
        return pd.Series([adjust, ldeladjust, rdeladjust], index=['adjust', 'leftDelAdjust', 'rightDelAdjust'])
    else:
        adjust = 0
        ldeladjust = LDel
        rdeladjust = RDel
        return pd.Series([adjust, ldeladjust, rdeladjust], index=['adjust', 'leftDelAdjust', 'rightDelAdjust'])

#function to draw deletion size distribution graph
def draw_deletion_size_distribution(df, sample_number, group, cut_site):
    deletion_df = df[df['IndelSize'] < 0]
    deletion_df = deletion_df.sort_values(by='IndelSize')
    deletion_df = deletion_df.reset_index(drop=True)    
    #adjust the start and end position when there's microhomology
    deletion_df["delstart"] = deletion_df['IndelPosition']
    deletion_df["delend"] = deletion_df['IndelPosition'] - deletion_df['IndelSize']
    adjust_condition = (deletion_df['IndelSize'] < 0) & (deletion_df['MHseq'].str.len() >= 1)
    deletion_df.loc[adjust_condition, 'delstart'] += deletion_df.loc[adjust_condition, 'MHseq'].str.len() // 2
    deletion_df.loc[adjust_condition, 'delend'] += deletion_df.loc[adjust_condition, 'MHseq'].str.len() // 2    
    y_values = range(len(deletion_df))
    conditions = [
        (deletion_df['IndelSize'] < 0 ) & (deletion_df['IndelSize'] >= -10),
        (deletion_df['IndelSize'] < -10) & (deletion_df['IndelSize'] >= -50),
        (deletion_df['IndelSize'] < -50) & (deletion_df['IndelSize'] >= -100),
        (deletion_df['IndelSize'] < -100)
    ]
    # Define the color for each condition
    colors = ['#53D2DC', '#3196E2', '#FF826C', '#FFC05F']
    # Apply the conditions to the 'color' column
    deletion_df['color'] = np.select(conditions, colors, default='black')  # 'black' is a default color if none of the conditions are met
    deletion_df['delstart'] = deletion_df['delstart'] - cut_site
    deletion_df['delend'] = deletion_df['delend'] - cut_site
    sns.set(style="whitegrid")
    '''graph for unadjust MH symmetry
    plt.figure(figsize=(10, 10))
    for y, (index, row) in zip(y_values, deletion_df.iterrows()):
        plt.hlines(y=y, xmin=row['IndelPosition'], xmax=row['IndelPosition'] - row['IndelSize'], 
                   color=row['color'], linewidth=0.5)
    cut_site_position = cut_site
    graph_lower_range = cut_site - 117
    graph_upper_range = cut_site + 117
    plt.axvline(x=cut_site_position, color='red', linestyle='--', linewidth=1)
    plt.xlabel('Reference Position')
    plt.ylabel('# of Read')
    graph_title = f'Deletion Size Distribution {group}s{sample_number:02d} unadjust for MH'
    plt.title(graph_title)
    plt.xlim(graph_lower_range, graph_upper_range)
    plt.xticks(np.arange(graph_lower_range, graph_upper_range, 20))
    output_graph_file = f"deletionSizeDist.unadjustMH.{group}s{sample_number:02d}.png"
    plt.savefig(output_graph_file, dpi=300)
    plt.show()
    '''
    title_fontsize = 14
    axis_label_fontsize = 12
    tick_label_fontsize = 10
    plt.figure(figsize=(10, 10))
    for y, (index, row) in zip(y_values, deletion_df.iterrows()):
        plt.hlines(y=y, xmin=row['delstart'], xmax=row['delend'], color=row['color'], linewidth=0.5)
    plt.axvline(x=0, color='red', linestyle='--', linewidth=1)
    plt.xlabel('Reference Position', fontsize = axis_label_fontsize)
    plt.ylabel('Number of Reads', fontsize = axis_label_fontsize)
    graph_title = f"Deletion Size Distribution {group}s{sample_number:02d} adjust for MH"
    plt.title(graph_title, fontsize = title_fontsize)
    plt.xlim(-150, 150)
    plt.xticks(np.arange(-150, 150, 20), fontsize = tick_label_fontsize)
    plt.yticks(fontsize = tick_label_fontsize)
    output_graph_file = f"deletionSizeDist.adjustMH.{group}s{sample_number:02d}.png"
    plt.savefig(output_graph_file, dpi=300)

def classify_indel(row, cutSite):
    if pd.notna(row['IndelSize']):
        if row['IndelSize'] < 0:
            row['LeftDeletion'] = cutSite - row['IndelPosition']
            row['RightDeletion'] = abs(row['IndelSize']) - row['LeftDeletion']
            #row['type'] = 'deletion'
        elif row['IndelSize'] > 0:
            row['LeftDeletion'] = 0
            row['RightDeletion'] = 0
            #row['type'] = 'insertion'
        elif row['IndelSize'] == 0:
            row['LeftDeletion'] = 0
            row['RightDeletion'] = 0
           # row['type'] = 'parental'
    else:
        row['LeftDeletion'] = np.nan
        row['RightDeletion'] = np.nan
        #row['type'] = np.nan
    return row

#########################################################################################################################
def findMicrohomology(ref, read, alignStart, indelStart, indelSize, cutSite):
    if pd.isna(indelStart) or pd.isna(indelSize):
        return pd.Series({'MHseq': pd.NA, 'Side': pd.NA, 'Type': pd.NA})
    elif indelSize == 0:
        return pd.Series({'MHseq': pd.NA, 'Side': pd.NA, 'Type': "parental"})
    elif indelSize > 0:
        return pd.Series({'MHseq': pd.NA, 'Side': pd.NA, 'Type': "insertion"})
    else:
        # Deletion
        deletionSize = abs(indelSize)
        deletionSeq = ref[indelStart - 1:indelStart + deletionSize - 1]
        refAfterIndel = ref[indelStart + deletionSize - 1:indelStart + deletionSize + 2 * deletionSize]
        refBeforeIndel = ref[indelStart - deletionSize:indelStart - 1]
        # Check for MH on the right side of the indel
        maxCommonLength = len(deletionSeq)
        commonSubstringRight = ""
        for i in range(1, maxCommonLength + 1):
            if deletionSeq[:i] == refAfterIndel[:i]:
                commonSubstringRight = refAfterIndel[:i]
            else:
                break
        # Check for MH on the left side of the indel
        commonSubstringLeft = ""
        for i in range(1, maxCommonLength + 1):
            if deletionSeq[-i:] == refBeforeIndel[-i:]:
                commonSubstringLeft = refBeforeIndel[-i:]
            else:
                break
        # Choose the longest MH
        if len(commonSubstringRight) == 0 and len(commonSubstringLeft) == 0:
            return pd.Series({'MHseq': pd.NA, 'Side': pd.NA, 'Type': "EJ"})
        elif len(commonSubstringRight) >= len(commonSubstringLeft):
            return pd.Series({'MHseq': commonSubstringRight, 'Side': "R", 'Type': "MMEJ"})
        else:
            return pd.Series({'MHseq': commonSubstringLeft, 'Side': "L", 'Type': "MMEJ"})

def get_additional_info(group):
    most_freq_sequence = group['Sequence'].value_counts().idxmax()
    row = group[group['Sequence'] == most_freq_sequence].iloc[0]
    return pd.Series({
        'Seq': most_freq_sequence,
        'IndelPosition': row['IndelPosition'],
        'IndelSize': row['IndelSize'],
        'LeftDeletion': row['LeftDeletion'],
        'RightDeletion': row['RightDeletion'],
        'MHseq': row['MHseq'],
        'Side': row['Side'],
        'Type': row['Type']
        })

#########################################################################################################################
```

```python

if len(sys.argv) < 5:
    print("Usage: python script.py <file_name_pattern> <group_number> <cut_site> <ref_file>")
    sys.exit(1)
file_name_pattern = sys.argv[1]
group = sys.argv[2]
cutSite = int(sys.argv[3])
ref = sys.argv[4]               
ref = getRef_fasta(ref)
if group == 0:
    group = ""
else:
    group = "g" + group
# Generate file names based on the pattern
match = re.match(r"(.+)\{(\d+):(\d+)\}(.+)", file_name_pattern)
if not match:
    print("Invalid file name pattern. Please use the format: prefix{start:end}suffix")
    sys.exit(1)
prefix, start, end, suffix = match.groups()
file_names = [f"{prefix}{i:02d}{suffix}" for i in range(int(start), int(end) + 1)]
for i, file_name in enumerate(file_names, start=int(start)):
    print(f"Processing {file_name}")
    df = read_sam_file(file_name)
    df[['IndelPosition', 'IndelSize']] = df.apply(lambda row: pd.Series(find_indel_positions_and_sizes(row['CIGAR'], row['StartPosition'], cutSite, row['Sequence'], ref)), axis=1)
    df['IndelPosition'] = df['IndelPosition'].astype('Int64')
    df['IndelSize'] = df['IndelSize'].astype('Int64')
    df = df.apply(classify_indel, axis=1, cutSite=cutSite)
    df[['MHseq', 'Side', 'Type']] = df.apply(lambda row: findMicrohomology(ref, row['Sequence'], row['StartPosition'], 
                                                                                       row['IndelPosition'], 
                                                                                       row['IndelSize'], 
                                                                                       cutSite),
                                                         axis=1, result_type='expand')
    df['LeftDeletion'] = df['LeftDeletion'].astype('Int64')
    df['RightDeletion'] = df['RightDeletion'].astype('Int64')
    df['MHsize'] = df['MHseq'].str.len()
    df[['adjust', 'leftDelAdjust', 'rightDelAdjust']] = df.apply(lambda row: adjust_MHposition(row['IndelSize'], row['MHsize'],
                                                                               row['LeftDeletion'], row['RightDeletion']),
                                                         axis=1, result_type='expand')
        
    print(f"Processing indel size pct for {file_name}")
    indelSizePct_df = calculate_indel_size_percentage(df)
    indelSizePct_file = f"indelSizePct.{group}s{i:02d}.csv"
    indelSizePct_df.to_csv(indelSizePct_file,index=False)
    
    print(f"Processing freq table for {file_name}")
    #generate frequency table
    cigar_count = df['CIGAR'].value_counts().reset_index()
    cigar_count.columns = ['CIGAR', 'Count']
    total_count = cigar_count['Count'].sum()
    cigar_count['pct'] = (cigar_count['Count'] / total_count) * 100
    cigar_count['pct'] = cigar_count['pct'].round(2)
    # For each unique CIGAR, find the row with the most frequent sequence and extract additional information
    additional_info = df.groupby('CIGAR').apply(get_additional_info).reset_index()
    freq_df = pd.merge(cigar_count, additional_info, on='CIGAR')
    freq_df_file = f"freq.{group}s{i:02d}.csv"
    freq_df.to_csv(freq_df_file,index=False)

    # Draw and save the deletion size distribution graph
    print(f"Processing deletion size distribution for {file_name}")
    draw_deletion_size_distribution(df, i, group, cutSite)
    
    print(f"Output written for {file_name}")
    rawData_file = f"{group}s{i:02d}.rawData.csv"
    df.to_csv(rawData_file, index=False)
    print(f"Finish {file_name}")
```

```python
import pysam
from collections import Counter

# Open the SAM file
samfile = pysam.AlignmentFile("test.sam", "r")

# Initialize a counter to store the frequency of each CIGAR score
cigar_counter = Counter()

# Iterate over each alignment in the SAM file
for alignment in samfile:
    # Get the CIGAR string for the alignment
    cigar = alignment.cigarstring
    
    # Increment the count for this CIGAR score
    cigar_counter[cigar] += 1

# Close the SAM file
samfile.close()

# Sort the frequency table by descending order of counts
sorted_cigar_counts = cigar_counter.most_common()

# Print the sorted frequency table
for cigar, count in sorted_cigar_counts:
    print(f"CIGAR: {cigar}\tCount: {count}")
```
